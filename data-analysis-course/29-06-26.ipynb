# PANDAS - Read & Manipulate one or multiple excel sheets.

```python
import pandas as pd
```

```python
print("\n ========= Read First Sheet of Excel File ============")

df = pd.read_excel("student.xlsx")
print(df)
```

### Read Specific Sheet

```python
sheet_2 = pd.read_excel("student.xlsx", sheet_name="Sheet2")
print(sheet_2)
```

### Read Sheet By Index

```python
print("\n=============== Read 2nd Sheet by its Index ==================")

sheet_2 = pd.read_excel("student.xlsx", sheet_name=1)
print(sheet_2)
```

### Read Multiple Sheet at One Time

```python
multiple_sheet = pd.read_excel("student.xlsx", sheet_name=[0, 1])
print(multiple_sheet)
```

### Read All Sheets

```python
all_sheets = pd.read_excel("student.xlsx", sheet_name=None)

for sheet_name, data in all_sheets.items():
    print("\nSheet Name:", sheet_name)
    print("-------------------------------------------------")
    print(data)
```

### Select Data

```python
print("\n===== Single Column =====")
print(df["name"])

print("\n===== Multiple Columns =====")
print(df[["name", "marks"]])

print("\n===== First Row =====")
print(df.iloc[0])
```

### Missing Values

```python
print("\n===== Missing Values =====")

print(df.isnull().sum())

df["marks"] = df["marks"].fillna(df["marks"].mean())
df["city"] = df["city"].fillna('Unknown')

print(df)
```

### Remove Duplicate Record

```python
print("\nDuplicate Rows: ")

df.duplicated().sum()
df = df.drop_duplicates()
print("\n===== After Removing Duplicates =====")
print(df)
```
