# PANDAS - Read & Manipulate one or multiple excel sheets.

```python
import pandas as pd
```

```python
print("\n ========= Read First Sheet of Excel File ============")

df = pd.read_excel("student.xlsx")
print(df)
```

### Read Specific Sheet

```python
sheet_2 = pd.read_excel("student.xlsx", sheet_name="Sheet2")
print(sheet_2)
```

### Read Sheet By Index

```python
print("\n=============== Read 2nd Sheet by its Index ==================")

sheet_2 = pd.read_excel("student.xlsx", sheet_name=1)
print(sheet_2)
```

### Read Multiple Sheet at One Time

```python
multiple_sheet = pd.read_excel("student.xlsx", sheet_name=[0, 1])
print(multiple_sheet)
```

### Read All Sheets

```python
all_sheets = pd.read_excel("student.xlsx", sheet_name=None)

for sheet_name, data in all_sheets.items():
    print("\nSheet Name:", sheet_name)
    print("-------------------------------------------------")
    print(data)
```

### Select Data

```python
print("\n===== Single Column =====")
print(df["name"])

print("\n===== Multiple Columns =====")
print(df[["name", "marks"]])

print("\n===== First Row =====")
print(df.iloc[0])
```

### Missing Values

```python
print("\n===== Missing Values =====")

print(df.isnull().sum())

df["marks"] = df["marks"].fillna(df["marks"].mean())
df["city"] = df["city"].fillna('Unknown')

print(df)
```

### Remove Duplicate Record

```python
print("\nDuplicate Rows: ")

df.duplicated().sum()
df = df.drop_duplicates()
print("\n===== After Removing Duplicates =====")
print(df)
```

#### Change Data Type

```python
print(df.dtypes)

df["age"] = df["age"].astype(str)

print("\nAfter changing datatype of age column:", df.dtypes)
```

#### Date & Time 

```python
df["joining_date"] = pd.to_datetime(df["joining_date"])  # here converting joining_date column in to real date

df["year"] = df["joining_date"].dt.year     # to get year from that date

df["month"] = df["joining_date"].dt.month_name()    # to get month name from that date

df["day"] = df["joining_date"].dt.day_name()    # to get day name form that date

print(df[["joining_date", "year", "month", "day"]])
```

#### Filter Data

```python
print("\n========= Filter Marks > 80 ===========")

result = df[df["marks"] > 80]
print(result)

print("\n========= Filter City =================")

city = df[df["city"] == "Valsad"]
print(city)
```

#### Sorting Data

```python
print("\n============== Ascending Order ==============")

result = df.sort_values("marks")
print(result)

print("\n============== Descending Order =============")

result = df.sort_values("marks", ascending=False)
print(result)
```

#### Add new column

```python
df["result"] = df["marks"].apply(lambda x: "Pass" if x >= 35 else "Fail")

print("\n----------------------------------------------------------")
print(df)
```

#### Rename Column

```python
df.rename(
    columns={
        "name": "student_name"
    },
    inplace=True
)

print(df.columns)
```

#### Delete Column

```python
df.drop("age", axis=1, inplace=True)

print(df)
```

#### Save Updated Data into Excel File

```python
df.to_excel("student.xlsx", index=False)

print("\nFile Update Successfully")
```

# PANDAS - Sorting, Filtering, Aggregation, Merge And Join 

```python
student_data = {
    "name": ["Raj", "Ved", "Dixit", "Vivek"],
    "id": [101, 102, 103, 104],
    "age": [23, 22, 21, 28],
    "marks": [89, 95, 78, None],
    "city": ["Suart", "Bharuch", "Valsad", None],
    "course": ["Python", "AI/ML", "Python", "Node JS"],
    "joining_date": ["15-06-2026", "20-05-2025", "01-02-2024", "01-12-2020"]
}

df = pd.DataFrame(student_data)
print(df)
```

### Aggregation

```python
print("\nAverage marks Corse wise")
print(df.groupby("course")["marks"].mean())

print("\nMaximum marks course wise")
print(df.groupby("course")["marks"].max())

print("\nMinimum marks course wise")
print(df.groupby("course")["marks"].min())

print("\nCourse wise students")
print(df.groupby("course")["id"].count())
```

#### Merge Data

```python
fees = {
    "id": [101, 102, 103, 104],
    "fees": [20000, 120000, 50000, 10000]
}

fee_df = pd.DataFrame(fees)
print(fee_df)

print("\n========== Inner Merge ==========")

inner = pd.merge(df, fee_df, on="id")
print(inner)
```
