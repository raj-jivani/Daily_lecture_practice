### Series in Python

##### A series is a one-dimensional labled array capable of holding any data type.

```python
import pandas as pd
```

```python
data = pd.Series([])
print(data)
```

```python
data = pd.Series([1,2,3,4,5])
print(data)
```

#### Series with custome index

```python
data = pd.Series(
    [90, 85, 95],
    index=["Math", "Phy", "Che"]
)
print(data)
```

#### Access Series Data

```python
print(data["Math"])
```

### DataFrame in Python

##### A DataFrame is a two-dimensional labled tublar structure considering of orws and columns.

```python
student = {
    "name": ["Rahul", "Raj", "Vivek"],
    "age": [21, 22, 23],
    "city": ["Surat", "Mumbai", "Pune"]
}

df = pd.DataFrame(student)
print(df)
```

#### Creation of DataFrame

```python
data = [
    ["Rahul", 21],
    ["Raj", 23],
    ["Vivek", 29]
]

df = pd.DataFrame(data)
print(df)
```

#### Exploring DataFrame

```python
student = {
    "name": ["Rahul", "Raj", "Vivek"],
    "age": [21, 22, 23],
    "city": ["Surat", "Mumbai", "Pune"]
}

df = pd.DataFrame(student)
```

#### Display last row

```python
print(df.tail(1))
```

#### Check Data Type

```python
print(df.dtypes)
```

#### Shape of DataFrame

```python
print(df.shape) # it returns number of rows and columns.
print(df.columns) # it returns name of all columns.
```

#### Statistical Info

```python
print(df.describe())
print(df.info())
```

## Operation on DataFrame

#### Select single column

```python
print(df["name"])
```

#### Select multiple column

```python
print(df[["name", "age"]])
```

#### Add new column

```python
df["grade"] = ["a", "b", "c"]
print(df)
```

#### Update column

```python
df["age"] = df["age"] + 5
print(df)
```

#### Delete column

```python
df.drop("grade", axis=1, inplace=True)
print(df)
```

#### Filter rows

```python
result = df[df["age"] > 22]
print(result)
```

#### Sorting Data
1. Ascending
2. Decending

```python
print(df.sort_values("age"))                    # Ascending Order
print(df.sort_values("age", ascending=False))   # Decending Order

print(df["age"].max())  # return maximum age
print(df["age"].min())  # return minimum age
print(df["age"].mean()) # return mean value of age
print(df["age"].sum())  # return sum of all age value
```

## Pandas Series All Data types with NaN values

#### What is NaN?

##### NaN[Not a Number] represnt missing or undfined value in pandas

```python
import numpy as np
```

```python
s = pd.Series([10, 20, np.nan, 40, 50])
print(s)
```

#### String Series with NaN value

```python
s = pd.Series(["Rahul", "Raj", np.nan, "Vivek"])
print(s)
```

#### Boolean Series with NaN

```python
s = pd.Series([True, False, np.nan, False, True])
print(s)
```

#### Mix Data type Series with NaN

```python
s = pd.Series([True, 23, "raj", np.nan, 12.12])
print(s)
```

#### Series with None

```python
s = pd.Series([10, None, 30, 40])
print(s)
print(s.isnull())
print(s.notnull())
print(s.isnull().sum())
```

#### Fill NaN value

##### using fillna()

```python
s = pd.Series([10, None, 30, 40])
s = s.fillna(20)
print(s)
```

##### using dropna()

```python
s = pd.Series([10, None, 20, 30])
s = s.dropna()
print(s)
```
