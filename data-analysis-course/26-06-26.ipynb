# Create, Read, and Write CSV, Excel & JSON File.

```python
import pandas as pd
```

### Create Dictionary

```python
student_data = {
    "name": ["Raj", "Ved", "Dixit", "Vivek"],
    "age": [23, 22, 21, 28],
    "marks": [89, 95, 78, None],
    "city": ["Suart", "Bharuch", "Valsad", None],
    "joining_date": ["15-06-2026", "20-05-2025", "01-02-2024", "01-12-2020"]
}

df = pd.DataFrame(student_data)
print(df)
```

#### to_csv

```python
df.to_csv("student.csv", index=False)
print("\n CSV File created Successfully.")
```

#### to_excel

```python
df.to_excel("student.xlsx", index=False)
print("\n Excel File Created Successfully.")
```

#### JSON

```python
df.to_json("studentData.json", orient="records", index=4)
print("\n JSON File Created Successfully.")
```

#### Read CSV File

```python
csv_data = pd.read_csv("student.csv")
print("\n", csv_data)
```

#### Basic Information

```python
print("\n ===== Shape =====")
print(csv_data.shape)

print("\n ===== Columns =====")
print(csv_data.columns)

print("\n ===== Data-Type =====")
print(csv_data.dtypes)

print("\n ===== Information =====")
print(csv_data.info())

print("\n ====== Describe =====")
print(csv_data.describe())
```
